```python
import pandas as pd
# 各模型 RMSE 比較
results = pd.DataFrame({
    "Model": [
        "XGB Random Search",
        "XGBoost v2",
        "Lasso",
        "Ridge",
        "Random Forest"
    ],
    "RMSE": [
        0.11158,
        0.11167,
        0.11175,
        0.11273,
        0.13212
    ]
})

results = results.sort_values(
    by="RMSE",
    ascending=True
)

print("各模型 RMSE：")
display(results)

best_model = results.iloc[0]

print(
    f"\n最佳模型：{best_model['Model']}"
)

print(
    f"RMSE：{best_model['RMSE']:.5f}"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    results["Model"],
    results["RMSE"]
)

plt.title("Model Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")

plt.xticks(rotation=15)

plt.tight_layout()

plt.show()
```

```python
ensemble_results = pd.DataFrame({
    "Weight (Lasso : XGB Random Search)": [
        "58 : 42",
        "62 : 38",
        "64 : 36",
        "66 : 34",
        "68 : 32"
    ],
    "Public Score": [
        0.12309,
        0.12311,
        0.12285,
        0.12317,
        0.12322
    ]
})

ensemble_results = ensemble_results.sort_values(
    by="Public Score"
)

print("Ensemble 比例測試：")
display(ensemble_results)

best = ensemble_results.iloc[0]

print(
    f"\n最佳 Ensemble：{best['Weight (Lasso : XGB Random Search)']}"
)

print(
    f"Public Score：{best['Public Score']:.5f}"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

plt.bar(
    ensemble_results["Weight (Lasso : XGB Random Search)"],
    ensemble_results["Public Score"]
)

plt.title("Ensemble Weight Comparison")
plt.xlabel("Weight")
plt.ylabel("Public Score")

plt.tight_layout()

plt.show()
```
