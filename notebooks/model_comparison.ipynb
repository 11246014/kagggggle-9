```python
import pandas as pd
```

```python
results = pd.DataFrame({
    "Model": [
        "XGB Random Search",
        "XGBoost v2",
        "Lasso",
        "Ridge",
        "Random Forest"
    ],
    "RMSE": [
        0.11158,
        0.11167,
        0.11175,
        0.11273,
        0.13212
    ]
})
```

```python
results = results.sort_values(
    by="RMSE",
    ascending=True
)

results
```

```python
best_model = results.iloc[0]

print(
    f"最佳模型：{best_model['Model']}"
)

print(
    f"RMSE：{best_model['RMSE']:.5f}"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    results["Model"],
    results["RMSE"]
)

plt.title("Model Comparison")
plt.ylabel("RMSE")

plt.show()
```
