```python
import pandas as pd
```

```python
results = pd.DataFrame({
    "Model": [
        "Ridge",
        "Lasso",
        "Random Forest",
        "XGBoost"
    ],
    "RMSE": [
        0.11525,
        0.11359,
        0.13802,
        0.11963
    ]
})

results
```

```python
results = results.sort_values(
    by="RMSE",
    ascending=True
)

results
```

```python
best_model = results.iloc[0]

print(
    f"最佳模型：{best_model['Model']}"
)

print(
    f"RMSE：{best_model['RMSE']:.5f}"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    results["Model"],
    results["RMSE"]
)

plt.title("Model Comparison")
plt.ylabel("RMSE")

plt.show()
```
